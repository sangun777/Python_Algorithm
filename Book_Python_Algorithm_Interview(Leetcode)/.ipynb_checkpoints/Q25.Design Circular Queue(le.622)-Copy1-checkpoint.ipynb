## 원형 큐를 구현하라!

* 큐와 같은 FIFO(선입선출) 구조
* 마지막 위치는 첫번째 위치와 연결된다.
### 큐의 경우에는 앞에 썼던 메모리 공간이 dequeue로 비워져도 다시 사용할 수 없지만,
### 원형 큐에 경우에는 앞쪽에 공간이 남아있다면 사용 가능하다!
  
  
* 같은 말로 Ring Buffer라 부른다.
* 프로그래밍 언어에 내장된 큐를 사용하지 말고, 풀 것!

## 원형 큐(=환형 큐 = Ring Buffer) 개념 정리

* front = rear = 0 으로 초기화
* front = rear는 큐가 공백이라는 의미
* 값이 하나 추가(enqueue)될 때마다, rear+=1, 
* 값이 하나 삭제+반환(dequeue)될 때마다, front+=1
* (rear+1)%q_size == front라면, 포화 상태로 간주
    * 원형 큐는 공백과 포화 구분을 위해 실제 사이즈 보다, 1개 요소가 덜 들어간다.
        * [0], [1], [2], [3] 인덱스를 같는 크기가 4인 원형 큐는 요소가 3개 들어갔을 때, 포화 상태로 인식한다.
            * 시작은 front = rear = 0 이며,
            * 값이 3개가 들어가면, rear = 3, front = 0이 되어, 
            * '(rear+1)%size == front'  => '4%4 ==0' 이 된다.

 * 초기화 : front = rear = 0
     * front == rear는 원형 큐가 공백상태라는 의미
     * 원형 큐의 포화판단을 위해 0으로 초기화
 
 * Enqueue
     * rear를 +1하고, (+1된)rear위치에 값 추가
     * 배열의 포화 상태 판단을 위해서,
         * (rear+1)%size = front 라면 포화 상태로 판단하여, 데이터 삽입이 이루어지지 않음
         * 예시) 
         
 * Dequeue
     * front를 +1하고, 그 위치에 있는 값을 가져오기
     * front == rear면 공백으로 판단하여 실행되지 않음
     * 해당 문제에서 dequeue는 삭제만 수행하지만, 일반적으로 dequeue는 삭제+반환을 수행한다.
         * 출처 : Introduction to Algorithms

# 내 구현

```python
class MyCircularQueue:
    
    # k는 원형 큐에 들어갈 수 있는 자료 갯수
    def __init__(self, k: int):
        self.front, self.rear = 0, 0
        self.cq = [None]
        self.size = k+1
        

    def enQueue(self, value: int) -> bool:
        if not (self.rear+1) % self.size == self.front:
            self.rear += 1
            self.cq.append(value)
            return True
        else:
            return False

    # 여기서 dequeue는 값을 반환하지 않음    
    def deQueue(self) -> bool:
        if self.front == self.rear:
            return False
        else:
            self.front += 1
            return True

    # 맨 앞 자료 반환, 비워져있으면 -1반환
    def Front(self) -> int:
        if self.front == self.rear:
            return -1
        else:
            return self.cq[self.front+1]
        
    # 마지막에 들어간 자료 반환, 비워져있으면 -1반환
    def Rear(self) -> int:
        if self.front == self.rear:
            return -1
        else:
            return self.cq[-1]
        
    def isEmpty(self) -> bool:
        if self.front == self.rear:
            return True
        else:
            return False
        
    def isFull(self) -> bool:
        if (self.rear+1)%self.size == self.front:
            return True
        else:
            return False
        

# Your MyCircularQueue object will be instantiated and called as such:
# obj = MyCircularQueue(k)
# param_1 = obj.enQueue(value)
# param_2 = obj.deQueue()
# param_3 = obj.Front()
# param_4 = obj.Rear()
# param_5 = obj.isEmpty()
# param_6 = obj.isFull()
```
