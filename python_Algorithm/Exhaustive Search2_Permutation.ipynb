# 순열 로직

1. A[i-1] <= A[i]를 만족하는 i 중 가장 큰 값을 찾는다.
   (커지는 부분 중 가장 뒷 부분)
   => 여기서 i-1 뒤에는 전부 내림차순이 됨
   => 실질적으로는 뒤에서 부터 내림차순이 아닌 가장 마지막 부분(i-1)을 찾는 것 임!
    => 여기서는 i=3
2. j >= i 중, A[j] >  A[i-1] 을 만족하는 가장 큰 j의 값을 찾는다.
   (A[i-1]보다 큰 값 중에 가장 작은 값의 인덱스 찾기)
    * 여기서 i-1은 2 이므로, A[2]=3보다 큰 j값 찾기
    => 여기서 A[j]=4, j=5
    
3. A[i-1]과 A[j]를 Swap 한다,
    => 여기서는 3과 4의 위치를 바꾼다.
    이전 : [7, 2, 3, 6, 5, 4, 1]
    결과 : [7, 2, 4, 6, 5, 3, 1]
4. i(를 포함한) 이후의 순열을 보두 뒤집는다.
    결과 : [7, 2, 4, 1, 3, 5, 6]

A = [7, 2, 3, 6, 5, 4, 1]

1. 여기서 보면 뒤에서 부터 내림차순 부분 6 5 4 1과 그 앞부분 7 2 3으로 나누어보면,
i는 6의 위치, i-1은 3의 위치(실질적으로 i-1 위치를 찾는 것)
2. 뒷부분의 6 5 4 1 중 3보다 큰 값 중 제일 뒤에 있는 값 4를 찾아서 서로 위치 Swap
3. 스왑하고 난 4의 뒷부분 6 5 3 1을 뒤집어 오름차순으로 만듬

만일, i가 제일 마지막 값이면, 그 앞의 값과 swap하는 결과 나오고 끝남

```python
# swap 함수 정의
def swap(arr, x, y):
    tmp = arr[x]
    arr[x] = arr[y]
    arr[y] = tmp
```

```python
def get_next_per(arr):
    n=len(arr)
    cnt = 0
    
    for i in range(1, n):
        if arr[i-1] <= arr[i]:
            temp = i-1 # 마지막 내림차순의 바로 앞부분
            cnt += 1 # 마지막 항 인지 확인하는 변수
            
    if cnt == 0: # 순열의 정렬 상태가 전부 내림차순 일 때,(마지막 항일 때)
        return
        
    for j in range(temp+1, n):
        if arr[j] >= arr[temp]:
            temp2 = j # 뒷부분에서 arr[temp] 보다 큰 값 중 가장 작은 값
    
    swap(arr, temp, temp2)
    
    temp3 = arr[temp+1:n]
    
    return arr[:temp+1] + temp3[::-1] # 뒷부분 뒤집기
```

```python
print(A)
A = get_next_per(A)
print(A)
```

```python
def get_next_per2(arr):
    n=len(arr)
    i = len(arr)-1 # 마지막 인덱스로 시작
    
    while(i > 0 and arr[i-1] > arr[i]):
        i-=1
    
    if i == 0: # 순열의 정렬 상태가 전부 내림차순 일 때,(마지막 항일 때)
        return
    
    temp = i-1          
        
    for j in range(temp+1, n):
        if arr[j] >= arr[temp]:
            temp2 = j # 뒷부분에서 arr[temp] 보다 큰 값 중 가장 작은 값
    
    swap(arr, temp, temp2)
    
    temp3 = arr[temp+1:n]
    
    return arr[:temp+1] + temp3[::-1] # 뒷부분 뒤집기
```
